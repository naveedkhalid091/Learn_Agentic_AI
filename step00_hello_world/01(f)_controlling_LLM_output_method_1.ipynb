## Controlling the output generated by the LLM:

We as a developer can control the LLM output by setting the **temprature** and the **max_output_tokens** range.

- **Temprature:** This parameter controls the creativity of the response, the temprature range would normally **from 0 to 1**. Bigger the number (i.e. 1) will make the LLM more creative and shorter range will make the LLM more specific (straightforward answer) & less creative.  
- **max_output_tokens:** This parameter set the LLM limit of the maximum number of tokens in a single prompt.

```python
import google.generativeai as genai
```

```python
from google.colab import userdata
GOOGLE_API_KEY=userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)
```

```python
model=genai.GenerativeModel(
    "gemini-2.0-flash-exp",
     generation_config=genai.GenerationConfig(
     temperature=0.7,
     max_output_tokens=10,
     )
)
```

```python
response=model.generate_content("How is Imran Khan? write 50 lines for him ")

print(response.text)
```

**Note:** Even I have requested to write 50 lines but LLM has stepped generating content after 22 tokens.

```python
model=genai.GenerativeModel(
    "gemini-2.0-flash-exp",
     generation_config=genai.GenerationConfig(
     temperature=0.7,
     max_output_tokens=1000,
     )
)
```

```python
response=model.generate_content("How is Imran Khan? write 50 lines for him ")

print(response.text)
```
