```python
import pymupdf4llm
import fitz  # PyMuPDF
import pytesseract
from PIL import Image
import io
import pathlib

pdf_path = "/content/Chemistry-9 complete-2025-26.pdf"

# Attempt to convert the PDF to Markdown using pymupdf4llm
md_text = pymupdf4llm.to_markdown(pdf_path)

# Define a simple heuristic: if the extracted text is very short or contains many dashes,
# assume that the conversion did not capture the actual book content.
if len(md_text.strip()) < 100 or md_text.count("-") > 50:
    print("Standard conversion produced insufficient content. Falling back to OCR extraction...")
    doc = fitz.open(pdf_path)
    ocr_pages = []
    for page_num in range(len(doc)):
        page = doc.load_page(page_num)
        # Render the page as an image at a higher DPI for better OCR accuracy
        pix = page.get_pixmap(dpi=300)
        img = Image.open(io.BytesIO(pix.tobytes()))
        # Extract text from the image using pytesseract
        text = pytesseract.image_to_string(img)
        ocr_pages.append(text)
    md_text = "\n\n".join(ocr_pages)
else:
    print("Standard conversion appears to have extracted sufficient content.")

# Save the final Markdown content to a file
output_file = pathlib.Path("complete_output_final.md")
output_file.write_bytes(md_text.encode("utf-8"))
print("Final Markdown conversion complete. Check 'complete_output_final.md'.")
```

```python
import pymupdf4llm
import pathlib

pdf_path = "/content/Chemistry-9 complete-2025-26.pdf"
# Convert PDF to Markdown with page_chunks enabled, which returns a list of dictionaries.
md_chunks = pymupdf4llm.to_markdown(pdf_path, page_chunks=True)

output_dir = pathlib.Path("markdown_chunks")
output_dir.mkdir(exist_ok=True)

# Save each chunk's text (accessed via the "text" key) as a separate Markdown file.
for i, chunk in enumerate(md_chunks):
    markdown_text = chunk.get("text", "")  # Extract the markdown text from the chunk dictionary.
    file_path = output_dir / f"page_{i+1}.md"
    file_path.write_bytes(markdown_text.encode("utf-8"))

# Optionally, combine all page texts into one complete Markdown file.
combined_markdown = "\n\n".join(chunk.get("text", "") for chunk in md_chunks)
pathlib.Path("complete_output.md").write_bytes(combined_markdown.encode("utf-8"))

print("Markdown conversion complete. Check the 'markdown_chunks' folder and 'complete_output.md'.")

```
